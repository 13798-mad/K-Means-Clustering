```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("data.csv")
data.head(5)
```

```python
from sklearn.cluster import KMeans
```

```python
sns.scatterplot(x="x",y="y",data=data)
```

```python
wcss=[]
clusters=range(1,10)
for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(data[["x","y"]])
    wcss.append(model.inertia_)
```

```python
plt.plot(clusters,wcss,"o-")
```

Number of clusters as per the Elbow methos is 3

```python
model=KMeans(n_clusters=3)
model.fit(data[["x","y"]])
```

```python
y_predict=model.predict(data[["x","y"]])
y_predict
```

```python
data["Group"]=y_predict
```

```python
data.head(5)
```

```python
plt.scatter(x=data["x"],y=data["y"],c=data["Group"])
plt.show()
```
