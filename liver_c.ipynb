```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("liver_cirrhosis.csv")
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data["Status"]=le.fit_transform(data["Status"])
data["Drug"]=le.fit_transform(data["Drug"])
data["Sex"]=le.fit_transform(data["Sex"])
data["Ascites"]=le.fit_transform(data["Ascites"])
data["Hepatomegaly"]=le.fit_transform(data["Hepatomegaly"])
data["Spiders"]=le.fit_transform(data["Spiders"])
data["Edema"]=le.fit_transform(data["Edema"])
```

```python
data.head(5)
```

```python
data.corr()
```

```python
x=data.drop(["Stage"],axis=1)
y=data["Stage"]
```

```python
sns.countplot(x="Stage",data=data)
plt.show()
```

```python
sns.heatmap(data.corr())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=49)
```

```python
rfr.fit(x_train,y_train)
```

```python
y_pred=rfr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r=r2_score(y_test,y_pred)
print(r*100)
```
